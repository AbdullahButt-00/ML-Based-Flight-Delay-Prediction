```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
Weather_Dataset = pd.read_csv('Data/weather_data.csv')
Weather_Dataset
```

```python
Training_Data = pd.read_csv('Data/merged_train.csv')
Training_Data
```

```python
Testing_Data = pd.read_csv('Data/merged_test.csv')
Testing_Data
```

```python
df1 = Training_Data
df2 = Weather_Dataset

months = []
days = []

for timestamp in df1['departure_estimated_runway']:
    if pd.notna(timestamp):
        try:
            timestamp = str(timestamp)
            dt_object = datetime.strptime(timestamp, "%Y-%m-%dT%H:%M:%S.%f")
            months.append(dt_object.strftime("%b"))
            days.append(dt_object.day)
        except ValueError:
            months.append(None)
            days.append(None)
    else:
        months.append(None)
        days.append(None)

df1['Month'] = months
df1['Day'] = days

for i in range(len(df1)):
    month = df1.at[i, 'Month']
    day = df1.at[i, 'Day']
    
    if pd.notna(month) and pd.notna(day):
        matching_row = df2[(df2['Month'] == month) & (df2['Day'] == day)]
        
        if not matching_row.empty:
            for col in df2.columns:
                if col not in ['Month', 'Day']:
                    df1.at[i, col] = matching_row.iloc[0][col]

Linked_Training_Dataset = df1
print("Data merged successfully into 'Linked_Training_Dataset'.")
```

```python
df = pd.read_csv('Data/Cleaned_Linked_Training.csv')
time_columns = ['departure_scheduled', 'departure_estimated_runway', 'departure_actual_runway', 'arrival_scheduled', 'arrival_estimated']

for col in time_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')
    
    df[f'{col}_Year'] = df[col].dt.year
    df[f'{col}_Month'] = df[col].dt.month
    df[f'{col}_Day'] = df[col].dt.day
    df[f'{col}_Time'] = df[col].dt.time

    df.drop(columns=[col], inplace=True)

df.to_csv('Data/Cleaned_Linked_Training.csv', index=False)

print("Timestamps processed and NaN values handled.")
```

```python
df = df[[col for col in df.columns if col != 'Delay in (hrs)'] + ['Delay in (hrs)']]

df.to_csv('Data/Cleaned_Linked_Training.csv', index=False)

print("Delay in (hrs) has been calculated and moved to the last column in the CSV file.")
```
