```python
import os
import json
from docx import Document
import pandas as pd

def convert_docx_to_json(docx_path):
    data = []
    doc = Document(docx_path)
    for para in doc.paragraphs:
        try:
            record = json.loads(para.text)
            data.append(record)
        except json.JSONDecodeError:
            continue
    return data

def flatten_json(nested_data):
    flat_list = []
    for item in nested_data:
        if isinstance(item, list):
            flat_list.extend(flatten_json(item))
        elif isinstance(item, dict):
            flat_list.append(item)
    return flat_list

def json_to_dataframe(data):
    flat_data = flatten_json(data)
    flattened_data = []
    for record in flat_data:
        if isinstance(record, dict):
            flattened_record = {
                "type": record.get("type"),
                "status": record.get("status"),
                "departure_iata": record.get("departure", {}).get("iataCode"),
                "departure_icao": record.get("departure", {}).get("icaoCode"),
                "departure_terminal": record.get("departure", {}).get("terminal"),
                "departure_scheduled": record.get("departure", {}).get("scheduledTime"),
                "departure_estimated_runway": record.get("departure", {}).get("estimatedRunway"),
                "departure_actual_runway": record.get("departure", {}).get("actualRunway"),
                "arrival_iata": record.get("arrival", {}).get("iataCode"),
                "arrival_icao": record.get("arrival", {}).get("icaoCode"),
                "arrival_terminal": record.get("arrival", {}).get("terminal"),
                "arrival_baggage": record.get("arrival", {}).get("baggage"),
                "arrival_scheduled": record.get("arrival", {}).get("scheduledTime"),
                "arrival_estimated": record.get("arrival", {}).get("estimatedTime"),
                "airline_name": record.get("airline", {}).get("name"),
                "airline_iata": record.get("airline", {}).get("iataCode"),
                "airline_icao": record.get("airline", {}).get("icaoCode"),
                "flight_number": record.get("flight", {}).get("number"),
                "flight_iata_number": record.get("flight", {}).get("iataNumber"),
                "flight_icao_number": record.get("flight", {}).get("icaoNumber"),
                "codeshare_airline_name": record.get("codeshared", {}).get("airline", {}).get("name") if record.get("codeshared") else None,
                "codeshare_airline_iata": record.get("codeshared", {}).get("airline", {}).get("iataCode") if record.get("codeshared") else None,
                "codeshare_airline_icao": record.get("codeshared", {}).get("airline", {}).get("icaoCode") if record.get("codeshared") else None,
                "codeshare_flight_number": record.get("codeshared", {}).get("flight", {}).get("number") if record.get("codeshared") else None,
                "codeshare_flight_iata_number": record.get("codeshared", {}).get("flight", {}).get("iataNumber") if record.get("codeshared") else None,
                "codeshare_flight_icao_number": record.get("codeshared", {}).get("flight", {}).get("icaoNumber") if record.get("codeshared") else None,
            }
            flattened_data.append(flattened_record)
    return pd.DataFrame(flattened_data)

input_dir = "/home/abdullah/Desktop/ML_Project/ML-Proj-Dataset/Train"
output_dir = "train_csv"

os.makedirs(output_dir, exist_ok=True)

for filename in os.listdir(input_dir):
    if filename.endswith(".docx"):
        docx_path = os.path.join(input_dir, filename)
        print(f"Processing file: {filename}")
        json_data = convert_docx_to_json(docx_path)
        df = json_to_dataframe(json_data)
        output_csv_path = os.path.join(output_dir, filename.replace(".docx", ".csv"))
        df.to_csv(output_csv_path, index=False)
        print(f"Saved CSV: {output_csv_path}")

print("All files processed and saved to 'train_csv' folder.")
```

```python
import pandas as pd

file_path = "/home/abdullah/Desktop/ML_Project/ML-Proj-Dataset/Weather/13.xlsx"
excel_data = pd.ExcelFile(file_path)
sheet1_data = excel_data.parse('Sheet1')

raw_data = sheet1_data.iloc[0]

def clean_and_split(data, group_size=3):
    values = [x for x in data if x not in ["Max", "Avg", "Min"]]
    return [values[i:i + group_size] for i in range(0, len(values), group_size)]

temperature_data = clean_and_split(raw_data['Temperature (°F)'].split())
dew_point_data = clean_and_split(raw_data['Dew Point (°F)'].split())
humidity_data = clean_and_split(raw_data['Humidity (%)'].split())
wind_speed_data = clean_and_split(raw_data['Wind Speed (mph)'].split())
pressure_data = clean_and_split(raw_data['Pressure (in)'].split())
precipitation_data = [x for x in raw_data['Precipitation (in)'].split() if x not in ["Total"]]

time = list(range(1, 32))  # Days 1 to 31
structured_data = pd.DataFrame({
    'Day': time,
    'Temperature Max (°F)': [float(t[0]) for t in temperature_data],
    'Temperature Avg (°F)': [float(t[1]) for t in temperature_data],
    'Temperature Min (°F)': [float(t[2]) for t in temperature_data],
    'Dew Point Max (°F)': [float(dp[0]) for dp in dew_point_data],
    'Dew Point Avg (°F)': [float(dp[1]) for dp in dew_point_data],
    'Dew Point Min (°F)': [float(dp[2]) for dp in dew_point_data],
    'Humidity Max (%)': [float(h[0]) for h in humidity_data],
    'Humidity Avg (%)': [float(h[1]) for h in humidity_data],
    'Humidity Min (%)': [float(h[2]) for h in humidity_data],
    'Wind Speed Max (mph)': [float(ws[0]) for ws in wind_speed_data],
    'Wind Speed Avg (mph)': [float(ws[1]) for ws in wind_speed_data],
    'Wind Speed Min (mph)': [float(ws[2]) for ws in wind_speed_data],
    'Pressure Max (in)': [float(p[0]) for p in pressure_data],
    'Pressure Avg (in)': [float(p[1]) for p in pressure_data],
    'Pressure Min (in)': [float(p[2]) for p in pressure_data],
    'Precipitation (in)': [float(p) for p in precipitation_data],
})

output_file_path = 'weather_data13.xlsx'
structured_data.to_excel(output_file_path, index=False)

print(f"Cleaned data saved to {output_file_path}")
```

```python
import os
import pandas as pd

input_dir = "weather_csv"  # Replace with your folder path
output_csv_path = "merged_weather_data.csv"  # Path for the final merged CSV

all_dataframes = []

for filename in os.listdir(input_dir):
    if filename.endswith(".xlsx"):
        file_path = os.path.join(input_dir, filename)
        print(f"Processing file: {filename}")
        
        df = pd.read_excel(file_path)
        
        df = df.rename(columns={
            "Temperature Max": "Temperature (°F) Max",
            "Temperature Avg": "Temperature (°F) Avg",
            "Temperature Min": "Temperature (°F) Min",
            "Dew Point Max": "Dew Point (°F) Max",
            "Dew Point Avg": "Dew Point (°F) Avg",
            "Dew Point Min": "Dew Point (°F) Min",
            "Humidity Max": "Humidity (%) Max",
            "Humidity Avg": "Humidity (%) Avg",
            "Humidity Min": "Humidity (%) Min",
            "Wind Speed Max": "Wind Speed (mph) Max",
            "Wind Speed Avg": "Wind Speed (mph) Avg",
            "Wind Speed Min": "Wind Speed (mph) Min",
            "Pressure Max": "Pressure (in) Max",
            "Pressure Avg": "Pressure (in) Avg",
            "Pressure Min": "Pressure (in) Min",
            "Precipitation": "Precipitation (in) Total",
            "Date": "Date",
            "Month": "Month",
            "Day": "Day"
        })

        df['Source File'] = filename
        
        all_dataframes.append(df)

merged_data = pd.concat(all_dataframes, ignore_index=True)

merged_data = merged_data.fillna(0)  

merged_data.to_csv(output_csv_path, index=False)

print(f"Merged and cleaned data saved to {output_csv_path}")
```
